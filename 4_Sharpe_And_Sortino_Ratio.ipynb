```python
"""
üìä Sharpe Ratio (Risk-Adjusted Return)
The Sharpe Ratio helps you understand how much return you're getting for each unit of risk you take.

Imagine you have two investments.

Both make money, but one goes up and down a lot (risky), and the other is more stable (less risky).

The Sharpe Ratio tells you which one gives you better returns relative to the risk.

Formula:

Sharpe¬†Ratio = (Average¬†Return ‚àí Risk-Free¬†Rate) / Standard¬†Deviation¬†of¬†Return


"""
```

```python
"""
üõ°Ô∏è Sortino Ratio (Downside-Risk Adjusted Return)
The Sortino Ratio is similar but focuses only on bad risk (when your investment drops in value).

It ignores good volatility (when returns go up).

It helps you see how much return you're getting for each unit of downside risk.

Formula:

Sortino¬†Ratio = (Average¬†Return ‚àí Risk-Free¬†Rate) / Downside¬†Deviation

"""
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#importing yahoo finance API to analyze a real stock
import yfinance as yf


# Simulate daily returns for 1 year (252 trading days)
np.random.seed(42)
"""daily_returns = np.random.normal(loc=0.001, scale=0.02, size=252)  # mean=0.1%, std dev=2%
dates = pd.date_range(start='2024-01-01', periods=252, freq='B')  # business days
returns = pd.Series(daily_returns, index=dates)
"""
"""
daily_returns = np.random.normal(loc=0.001, scale=0.04, size=252) #Adj std dev to 4%
dates = pd.date_range(start='2024-01-01', periods=252, freq='B')
returns = pd.Series(daily_returns,index=dates)
"""
#When increasing the standard deviation, the Sharpe and Sortino ratio both decrease, meaning that the relative performance of each stock has more risk (double) associated with its returns
#This logically follows as the standard deviation increases because this means that there is a greater overall volatility in the performance, and therefore a greater risk associated

"""
daily_returns = np.random.normal(loc=0.001, scale=0.005, size=252) #Adj std dev to 0.5%
dates = pd.date_range(start='2024-01-01', periods=252, freq='B')
returns = pd.Series(daily_returns,index=dates)

#Sharpe and sortino of 0.19 & 0.36 respectfully


daily_returns = np.random.normal(loc=0.001, scale=0.1, size=252) #Adj std dev to 10%
dates = pd.date_range(start='2024-01-01', periods=252, freq='B')
returns = pd.Series(daily_returns,index=dates)

#Sharpe and sortino of 0.1 each
"""

#The less the variance (and therefore volatility), the greater the Sharpe/Sortino ratio (and consequently cumulative ROI)



#Getting actual stock returns for Palantir from Aug 2024 through 2025

prices_df = yf.download("PLTR", start="2024-08-01", end="2025-08-01", interval="1d", auto_adjust=True)
prices = prices_df['Close']
daily_returns = prices.pct_change().dropna()

returns = pd.Series(daily_returns.to_numpy().squeeze())

#Palantir has had a great cumulative return on investment for the level of risk taken, with relatively high Sharpe and Sortino ratios


# Set risk-free rate (e.g., 0.02 annual, converted to daily)
risk_free_rate = 0.02 / 252

# Sharpe Ratio
excess_returns = returns - risk_free_rate
sharpe_ratio = excess_returns.mean() / returns.std()

# Sortino Ratio
downside_returns = returns[returns < 0]
downside_std = downside_returns.std()
sortino_ratio = excess_returns.mean() / downside_std

print(f"Sharpe Ratio: {sharpe_ratio:.2f}")
print(f"Sortino Ratio: {sortino_ratio:.2f}")

# Plot the returns
returns.cumsum().plot(title="Cumulative Return of Investment")
plt.show()
```
